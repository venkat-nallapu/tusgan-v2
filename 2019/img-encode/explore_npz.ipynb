# Explore NPZ File
This notebook explores the encoded data stored in the `.npz` file.

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Note: Update this path if you are trying to load a different npz file
npz_path = '../../wgan-gp/tusgan_encoded.npz'

if not os.path.exists(npz_path):
    print(f"File not found at: {npz_path}")
else:
    data = np.load(npz_path)
    print("Successfully loaded NPZ file.")
    print("\nKeys in the NPZ file:")
    for key in data.files:
        print(f" - {key}")
```

```python
if 'data' in locals():
    print("Shapes of the arrays:")
    for key in data.files:
        print(f"{key}: {data[key].shape} (dtype: {data[key].dtype})")
```

```python
if 'data' in locals() and 'diary_tensor' in data:
    # Extract a sample diary (first person in the dataset)
    # Shape is expected to be (N, 9, 48, 1)
    sample_idx = 0
    sample_diary = data['diary_tensor'][sample_idx]
    
    # Remove the last dimension for visualization -> shape (9, 48)
    sample_diary_2d = sample_diary.squeeze(-1)
    
    # Map the [-1, 1] range back to [0, 1] for easier visualization (since -1 was used for GAN)
    vis_diary = (sample_diary_2d + 1) / 2.0
    
    plt.figure(figsize=(15, 6))
    plt.imshow(vis_diary, aspect='auto', cmap='viridis')
    
    plt.title(f"Sample 24-Hour Activity Diary (Index: {sample_idx})")
    plt.xlabel("Time (Half-hours)")
    plt.ylabel("Major Division")
    plt.colorbar(label='Activity Presence (1=Active, 0=Inactive)')
    
    # Set y-axis labels for the 9 divisions
    plt.yticks(ticks=np.arange(9), labels=[
        '1: Employment', '2: Production', '3: Unpaid Domestic', '4: Unpaid Caregiving',
        '5: Unpaid Volunteer', '6: Learning', '7: Socializing/Religious',
        '8: Leisure/Sports', '9: Self-care'
    ])
    
    # Optional: Set x-axis labels to show actual times
    time_labels = [f"{str(h).zfill(2)}:00" for h in range(4, 28)] # 04:00 to 28:00 (04:00 next day)
    plt.xticks(ticks=np.arange(0, 48, 2), labels=time_labels, rotation=45)
    
    plt.tight_layout()
    plt.show()
```
